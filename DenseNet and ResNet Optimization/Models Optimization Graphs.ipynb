### ResNet Learning Rate vs Layers Graph

```python
rn_18_lr = [1e-2,1e-3,1e-4]
rn_18_TimeDelta = [1806.118 , 1807.222 , 1792.618]
rn_18_TimeDelta =[round(i/60,2) for i in rn_18_TimeDelta]

rn_18_TestAccuracy = [83.5,74.75 , 1.667 ]
rn_18_AveLoss =[0.470 ,0.593, 9.838 ]
```

```python
rn_50_lr = [1e-2,1e-3,1e-4]
rn_50_TimeDelta = [5432.991 , 5145.154 , 5442.251]
rn_50_TimeDelta =[round(i/60,2) for i in rn_50_TimeDelta]

rn_50_TestAccuracy = [88.25,81.17 , 4.917 ]
rn_50_AveLoss =[0.316 ,0.426, 7.275 ]
```

### DenseNet Learning Rate vs Layers Graph

```python
dn_121_lr = [1e-2,1e-3,1e-4]
dn_121_TimeDelta = [5108.736 , 5082.322 , 5072.685]
dn_121_TimeDelta =[round(i/60,2) for i in dn_121_TimeDelta]
dn_121_TestAccuracy = [85.92,78.08 , 3.08 ]
dn_121_AveLoss =[0.340 ,0.499,7.173 ]
```

```python
dn_169_lr = [1e-2,1e-3,1e-4]
dn_169_TimeDelta = [6790.120 , 6850.481 , 6748.484]
dn_169_TimeDelta =[round(i/60,2) for i in dn_169_TimeDelta]
dn_169_TestAccuracy = [85.5,78.67 , 11.33 ]
dn_169_AveLoss =[0.348 ,0.490, 4.883 ]
```

### Comparison Graphs

```python
### Time Delta vs LR

plt.figure(figsize=(12, 6))
plt.plot(rn_18_lr, rn_18_TimeDelta, label="ResNet 18", color='red', marker='o')
plt.plot(rn_50_lr, rn_50_TimeDelta, label="ResNet 50", color='green', marker='x')
plt.plot(dn_121_lr, dn_121_TimeDelta, label="DenseNet 121", color='blue', marker='s')
plt.plot(dn_169_lr, dn_169_TimeDelta, label="DenseNet 169", color='purple', marker='^')


plt.title("(Time Taken for Model Training & Testing) vs (Learning Rate)",fontsize = 16)
plt.xlabel('Learning Rate',fontsize = 16)
plt.ylabel("Time Taken (Hours)",fontsize = 16)
plt.legend()
plt.grid(True)
plt.show()
```

```python
### Accuracy vs LR

plt.figure(figsize=(12, 6))
plt.plot(rn_18_lr, rn_18_TestAccuracy, label="ResNet 18", color='red', marker='o')
plt.plot(rn_50_lr, rn_50_TestAccuracy, label="ResNet 50", color='green', marker='x')
plt.plot(dn_121_lr, dn_121_TestAccuracy, label="DenseNet 121", color='blue', marker='s')
plt.plot(dn_169_lr, dn_169_TestAccuracy, label="DenseNet 169", color='purple', marker='^')


plt.title("Model Test Accuracy vs Learning Rate",fontsize = 16)
plt.xlabel('Learning Rate',fontsize = 16)
plt.ylabel("Test Accuracy (%)",fontsize = 16)
plt.legend()
plt.grid(True)
plt.show()
```

```python
### Average Loss vs LR

plt.figure(figsize=(12, 6))
plt.plot(rn_18_lr, rn_18_AveLoss, label="ResNet 18", color='red', marker='o')
plt.plot(rn_50_lr, rn_50_AveLoss, label="ResNet 50", color='green', marker='x')
plt.plot(dn_121_lr, dn_121_AveLoss, label="DenseNet 121", color='blue', marker='s')
plt.plot(dn_169_lr, dn_169_AveLoss, label="DenseNet 169", color='purple', marker='^')


plt.title("Model Average Loss vs Learning Rate",fontsize = 16)
plt.xlabel('Learning Rate',fontsize = 16)
plt.ylabel("Average Loss",fontsize = 16)
plt.legend()
plt.grid(True)
plt.show()
```
